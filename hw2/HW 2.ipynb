```python
import numpy as np
import math                                       #Import required packages

s=np.arange(0,100,step=0.1)                       #Array of distance through cloud in pc
n = 1                                             #Particle density in cm^(-3)
I_0=float(input("Enter value of I\N{LATIN SUBSCRIPT SMALL LETTER V}(0): "))         #Intensity at s=0
S_v=float(input("Enter value of S\N{LATIN SUBSCRIPT SMALL LETTER V}: "))          #Source function
sigma_v0=float(input("Enter value of \N{GREEK SMALL LETTER SIGMA}\N{LATIN SUBSCRIPT SMALL LETTER V}\N{SUBSCRIPT ZERO}: "))    #Cross-section
I_v=np.zeros(len(s))                              #Define null vector for intensity
I_v[0]=I_0                                        #Define intensity at s=0
for i in range(0,len(s)):
    s_cm=s[i]*3.086*1e18                          #pc -> cm
    tau=s_cm*n*sigma_v0                           #Calculate optical depth at certain distance into cloud
    I_v[i]=I_0*np.exp(-tau)+S_v*np.exp(-tau)*(np.exp(tau)-1)  #Calculate intensity using RT eqn
print(I_v)




```

```python
def gaussian_sigma(sigma_v0,v1,v2,v_fwhm): #define a function that takes as input freq range, sigma_v0 and v_fwhm to generate Gaussian profile for sigma_v 
    sigma_v=[]
    v=np.arange(v1,v2,step=0.01)                     #array for freq interval
    v_0=(v1+v2)/2                          #peak position
    for i in range(0,len(v)):              #loop which uses a Gaussian fn with mean=v_0 and stdev=v_fwhm to give values of sigma_v at various freq
        c=0.95*sigma_v0*math.exp(-(v[i]-v_0)**2/v_fwhm**2)+0.05*sigma_v0 
        sigma_v=np.append(c,sigma_v)       #append value obtained to array
    return sigma_v                         #return array with values of sigma_v at different freq.
    
```

```python
import matplotlib.pyplot as plt

v1=7770                            #define freq range, sigma_v0, v_peak and v_fwhm
v2=7780
v_peak=(v1+v2)/2
v_fwhm=(v2-v1)/10
sigma_v0=3.24e-24
sigma_v=gaussian_sigma(sigma_v0,v1,v2,v_fwhm) #use previously defined function to generate gaussian profile
plt.figure(figsize=(9,6))               #figure size
plt.rc('xtick',labelsize=12)            #font for axes ticks and labels 
plt.rc('ytick',labelsize=12)
plt.rc('axes',labelsize=16)
plt.xlabel(r"Frequency ($\nu$) ($\AA$)")      #x and y axis labels     
plt.ylabel(r"Cross-section ($\sigma_\nu$) ($cm^2$)")
plt.ylim(0,3.3e-24)
plt.arrow(v_peak,0,0,0,lw=1,color='red')        #mark peak freq.
plt.annotate(r'$\nu_0$=7775 $\AA$',(v_peak+v_fwhm,3.0e-24),fontweight='light')    #annotate with value of peak freq. and peak cross section
plt.annotate(r'$\sigma_{\nu,0}=3.24*10^{-24}$ cm$^2$',(v_peak+v_fwhm,2.8e-24),fontweight='light')
plt.plot(np.arange(v1,v2,step=0.01),sigma_v)           #plot
```
