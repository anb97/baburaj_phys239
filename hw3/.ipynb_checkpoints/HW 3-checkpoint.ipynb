```python
import numpy as np

Z=2
x_i= 0; y_i=0          #ion position 
q_i=Z                 #ion charge in elementary charge
N=10000000                #no. of steps
dt=1e-14             #timestep duration
K=9*1e15             #electrostatic const.
x_e=np.zeros(N+1);y_e=np.zeros(N+1);v_x=np.zeros(N+1);v_y=np.zeros(N+1) #define null arrays for position, velocity & acceleration
a_x=np.zeros(N);a_y=np.zeros(N)
x_e[0]=-200000; y_e[0]=50000     #initial electron position in Bohr radii
q_e=1                    #electron charge in elementary charge
a0=5.29*1e-9             #Bohr radius in cm
e=1.602*1e-19                 #elementary charge in C
v_x[0]=1e7; v_y[0]=0      #initial electron velocity in cm/s
m_e=9.1*1e-31             #electron mass in kg
t=np.arange(0,N+1)        #define time array
```

```python
import matplotlib.pyplot as plt

for i in range(0,len(t)-1): 
    x_now=x_e[i]; y_now=y_e[i]                   #position at start of timestep
    r=((x_now)**2+(y_now)**2)**0.5               #distance from ion at start of timestep
    F=K*q_e*Z*e**2/(r*a0)**2                     #force acting during timestep
    F_x=-F*x_now/r;F_y=-F*y_now/r                #force along x and y
    a_x[i]=F_x/m_e                   #acceleration along x and y during timestep
    a_y[i]=F_y/m_e
    vx_new=v_x[i]+a_x[i]*dt                    #velocity along x and y at end of timestep
    vy_new=v_y[i]+a_y[i]*dt
    v_x[i+1]=vx_new                   #append new velocity to array 
    v_y[i+1]=vy_new
    x_now=x_now*a0+v_x[i]*dt          #update position at end of timestep
    y_now=y_now*a0+v_y[i]*dt 
    x_e[i+1]=x_now/a0                  #append new position to array
    y_e[i+1]=y_now/a0
t=t*dt                                #no. of steps array -> time array
    
    
    
```

```python
plt.figure(figsize=(9,6))               #figure size
plt.rc('xtick',labelsize=12)            #font size for axes ticks and axes labels
plt.rc('ytick',labelsize=12)
plt.rc('axes',labelsize=16)
plt.xlabel(r"X-distance from ion (Bohr radii $a_0$)")     #x and y axis labels        
plt.ylabel(r"Y-distance from ion (Bohr radii $a_0$)")
plt.title("Path of electron in x-y space",fontsize=16)    #title
plt.rc('legend',fontsize=12)            #font size for legend
plt.plot(0,0,'.',color='red')           #mark position of ion 
plt.annotate('Ion',(100,0))              #label ion
plt.xlim(1.1*x_e[0],-1.5*x_e[0])
plt.ylim(-y_e[0]-30000,y_e[0]+30000)
plt.plot(x_e,y_e,label='Path of electron')    
plt.legend()                            #show legend   
```

```python
plt.figure(figsize=(9,6))               
plt.rc('xtick',labelsize=12)             
plt.rc('ytick',labelsize=12)
plt.xlabel(r"Time (s)")        
plt.ylabel(r"Distance from ion (Bohr radii $a_0$)")
plt.title("X & Y position of electron wrt time",fontsize=16)  
plt.rc('legend',fontsize=12)          
plt.plot(t,x_e,label='Distance along x') 
plt.plot(t,y_e,label='Distance along y')
plt.legend()                                           
```

```python
plt.figure(figsize=(9,6))  
plt.rc('xtick',labelsize=12)            
plt.rc('ytick',labelsize=12)
plt.rc('axes',labelsize=16)
plt.title("Velocity of electron wrt time",fontsize=16)
plt.rc('legend',fontsize=12)
plt.xlabel(r"Time (s)")          
plt.ylabel(r"Velocity ($cm/s$)")
plt.xlim(0,1e-9)
plt.plot(t,v_x,label='Velocity along x')
plt.plot(t,v_y,label='Velocity along y')
plt.legend()
```

```python
plt.figure(figsize=(9,6))               
plt.rc('xtick',labelsize=12)            
plt.rc('ytick',labelsize=12)
plt.rc('axes',labelsize=16)
plt.title("Acceleration of electron wrt time",fontsize=16)
plt.rc('legend',fontsize=12)
plt.xlabel(r"Time (s)")          
plt.ylabel(r"Acceleration ($cm/s^2$)")
plt.xlim(0,0.5e-9)
plt.plot(t[1:N+1],a_x,label='Acceleration along x')
plt.plot(t[1:N+1],a_y,label='Acceleration along y')
plt.legend()
```

```python
power_spec_y=np.fft.rfft(a_y)                 #generate fourier transform
power_spec_y=np.abs(power_spec_y)**2          #power spectrum = abs(FT)^2
power_spec_x=np.fft.rfft(a_x)
power_spec_x=np.abs(power_spec_x)**2
freq=np.fft.fftfreq(power_spec_y.size,d=dt)  #generates the frequency array
plt.figure(figsize=(9,6))
plt.rc('xtick',labelsize=12)            
plt.rc('ytick',labelsize=12)
plt.rc('axes',labelsize=16)
plt.xlabel(r"Frequency (Hz)")          
plt.ylabel(r"Amplitude (arbitrary units)")
plt.title("Radiation power spectrum of electron-ion interaction")
plt.rc('legend',fontsize=10,loc='upper right')
o=freq.size                                 #we note size of freq array and plot only 0 & the positive half
plt.xlim(0,1e11)
plt.plot(freq[0:int((o+1)/2)],power_spec_x[0:int((o+1)/2)],'tab:blue',label='Radiation power spectrum for a_x')
plt.plot(freq[0:int((o+1)/2)],power_spec_y[0:int((o+1)/2)],'tab:orange',label='Radiation power spectrum for a_y')
plt.legend()
print(freq[np.argmax(power_spec_y)])        #print value of peak frequency
```

```python
a=(a_x**2+a_y**2)**0.5                         #magnitude of acceleration
power_spec=np.fft.rfft(a)
power_spec=np.abs(power_spec)**2
freq=np.fft.fftfreq(power_spec_y.size,d=dt)
plt.figure(figsize=(9,6))
plt.rc('xtick',labelsize=12)            
plt.rc('ytick',labelsize=12)
plt.rc('axes',labelsize=16)
plt.xlabel(r"Frequency (Hz)")          
plt.ylabel(r"Amplitude (arbitrary units)")
plt.title("Radiation power spectrum of electron-ion interaction")
plt.rc('legend',fontsize=10,loc='upper right')
plt.xlim(0,0.5e12)
plt.plot(freq[0:int((o+1)/2)],power_spec[0:int((o+1)/2)],'tab:blue',label='Radiation power spectrum for net acc.')
plt.legend()
```
